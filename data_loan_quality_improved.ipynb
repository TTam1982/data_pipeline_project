```python
import pandas as pd
import os

SOURCE_CSV_FILE = os.path.abspath("./datasets/cleaned/dataset-for-bank-loan-prediction_cleaned.csv")
# Đọc dữ liệu từ file CSV đã được clean
data = pd.read_csv(SOURCE_CSV_FILE)

# Tạo đặc trưng mới

# Tỷ lệ nợ hàng tháng so với thu nhập
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ nợ hàng tháng so với thu nhập")
data['debt_to_income_ratio'] = data['Monthly Debt'] / (data['Annual Income'] / 12)

# Số năm kinh nghiệm làm việc hiện tại
print("^^^^^^^^^^^^^^^^^^^ Số năm kinh nghiệm làm việc hiện tại")
data['years_in_current_job'] = data['Years in current job'].replace({'10+ years': 10, '< 1 year': 0.5}).str.extract('(\d+)').astype(float)

# Tỷ lệ số dư tín dụng hiện tại so với tín dụng tối đa
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ số dư tín dụng hiện tại so với tín dụng tối đa")
data['credit_utilization_ratio'] = data['Current Credit Balance'] / data['Maximum Open Credit']

# Số lượng tài khoản tín dụng trên mỗi năm lịch sử tín dụng
print("^^^^^^^^^^^^^^^^^^^ Số lượng tài khoản tín dụng trên mỗi năm lịch sử tín dụng")
data['accounts_per_year'] = data['Number of Open Accounts'] / data['Years of Credit History']

# Số năm kể từ lần vi phạm cuối cùng
print("^^^^^^^^^^^^^^^^^^^ Số năm kể từ lần vi phạm cuối cùng")
data['years_since_last_delinquent'] = data['Months since last delinquent'] / 12

# Số lượng vấn đề tín dụng trên mỗi năm lịch sử tín dụng
print("^^^^^^^^^^^^^^^^^^^ Số lượng vấn đề tín dụng trên mỗi năm lịch sử tín dụng")
data['credit_problems_per_year'] = data['Number of Credit Problems'] / data['Years of Credit History']

# Tỷ lệ số dư tín dụng hiện tại so với số dư tín dụng tối đa của từng năm lịch sử tín dụng
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ số dư tín dụng hiện tại so với số dư tín dụng tối đa của từng năm lịch sử tín dụng")
data['balance_to_max_credit_per_year'] = data['Current Credit Balance'] / (data['Maximum Open Credit'] * data['Years of Credit History'])

# Tỷ lệ phá sản (Bankruptcies) và các khoản thuế chưa thanh toán (Tax Liens)
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ phá sản (Bankruptcies) và các khoản thuế chưa thanh toán (Tax Liens)")
data['bankruptcies_per_year'] = data['Bankruptcies'] / data['Years of Credit History']
data['tax_liens_per_year'] = data['Tax Liens'] / data['Years of Credit History']

# Tỷ lệ nợ hàng tháng so với số dư tín dụng tối đa
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ nợ hàng tháng so với số dư tín dụng tối đa")
data['debt_to_max_credit_ratio'] = data['Monthly Debt'] / data['Maximum Open Credit']

# Tỷ lệ tài khoản mở (Open Accounts) so với số lượng vấn đề tín dụng (Credit Problems)
print("^^^^^^^^^^^^^^^^^^^ Tỷ lệ tài khoản mở (Open Accounts) so với số lượng vấn đề tín dụng (Credit Problems)")
data['open_accounts_to_credit_problems'] = data['Number of Open Accounts'] / (data['Number of Credit Problems'] + 1)  # +1 để tránh chia cho 0

# Hiển thị dữ liệu đã tạo thêm các đặc trưng mới
print("^^^^^^^^^^^^^^^^^^^ Hiển thị dữ liệu đã tạo thêm các đặc trưng mới")
print(data.head())

# Lưu dữ liệu đã được cải tiến vào một file CSV mới
ENHANCED_CSV_FILE = os.path.abspath("./datasets/cleaned/dataset-for-bank-loan-prediction_cleaned_with_enhanced.csv")
data.to_csv(ENHANCED_CSV_FILE, index=False)
print("Enhanced data has been saved to 'dataset-for-bank-loan-prediction_cleaned_with_enhanced'")
```
